## I. Export Sample Data from Mathematica

The nested data structure we were provided could be nicely stored/exported as JSON, but we must make a few changes, because Mathematica's JSON exporter expects a list of rules. To convert from the given Mathematica association to json, we can do the following:

```mathematica
[1] Clear[f]
    f[x_] := Table[ToString[i] -> ToString /@ x[[i]], {i, 10}]

[2] Export["coin_data.json", Normal[Map[f, data]]]
```

In line 1, we define the function `f[x_]` which takes a given students trail data (a list of 10 lists) and converts it to a set of "rules", with the trail number as a key. We must explicitly convert the trail number and the letters in the lists are all strings. For example:

```
{{H, H}, {H, T}} -> {"0" -> {"H", "H"}, "1" -> {"H", "T"}}
```

In line 2, we "map" this function to every value in `data`, and we also call `Normal` to convert the outer association structure into a set of rules. We can then write that to JSON, which we can import into Python using Pandas!

### Q1) Number of flips expected for given player to win?

#### Define Winner Function

First, let's define a function that can determine winner of a given trial, by looking at the final entry.

````python
def find_winner(lst):
    """Given list of flips, returns winner 'A' or 'B'
    
    If the final flip is heads, Player B has won
    If the final flip is tails, Player A has won
    """
    
    if lst[-1] == 'H':
        return 'B'
    else:
        return 'A'
````

````python
# Example usage

lst = data.iloc[0,0]  # results at first row and first column 
print('Trail :', lst)

winner = find_winner(lst) # returns 'A', since final entry is 'T'
print('Winner:', winner)
````

##### We can apply this function to every cell in the dataframe, using Pandas "applymap"

````python
data.applymap(find_winner)
````

##### We could also apply the built-in length function, to get the length of each list

````python
data.applymap(len)
````

#### Flatten Data

Backtrack. Since we don't care about trial number or person for now, let's flatten the Pandas dataframe into a Pandas series.

````python
flat_data = pd.Series(data.values.flatten())
````

````python
# use 'head' to see first 5 entries
flat_data.head()
````

````python
# find winner of each list, using apply

win_series = flat_data.apply(find_winner)
win_series.head()
````

````python
# find length of each list, using apply

len_series = flat_data.apply(len)
len_series.head()
````

#### Combine Data Back into Single DataFrame

We can make a dataframe from the two series we constructed

````python
df = pd.DataFrame({'winner':win_series,
                   'num_flips':len_series,})

# Show first 5 entries
df.head()
````

#### Finally, can filter DataFrame by winner and view stats

````python
# Filter DF down to just cells where 'A' won
a_wins = df.query("winner == 'A'")

# View some stats
a_wins.describe()
````

````python
# Filter DF down to just cells where 'B' won
b_wins = df.query("winner == 'B'")

# View some stats
b_wins.describe()
````
